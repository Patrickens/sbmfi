```python
def sample_ball(n, fcm, radius_root=None):
    K = len(fcm.net_theta_id())
    L = len(fcm.xch_theta_id())
    
    random_directions = random.normal(size=(K, n))
    random_directions /= linalg.norm(random_directions, axis=0)
    if radius_root is None:
        radius_root = K
    random_radii = random.random((n, 1)) ** (1.0 / radius_root)
    samples = (random_directions * random_radii[:,0]).T
    samples = [samples, random_radii]

    if L > 0:
        samples.append(random.random((n, L)))
    
    samples = np.hstack(samples)
    return pd.DataFrame(samples, columns=fcm.theta_id(coordinate_id='ball'))

biv_kwargs = dict(filled=True, colorbar=True)
```
