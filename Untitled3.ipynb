To compute the normalizing constant \( Z \) for a density \( f(x) \) supported on a polytope \( P \), we need to integrate \( f(x) \) over \( P \):

$$
Z = \int_P f(x) \, dx
$$

### **General Approaches**
Since $ P $ is a polytope, the integration must be constrained to its region. Here are some efficient ways to compute $ Z $:

#### **1. Monte Carlo Integration (Stochastic Approach)**
If direct integration is intractable, you can approximate $ Z $ using Monte Carlo sampling:
- Sample points \( x_1, x_2, \dots, x_N \) uniformly from $ P $.
- Compute the average of \( f(x) \).
- Multiply by the volume of \( P \), denoted as \( \text{Vol}(P) \).

$$
Z \approx \text{Vol}(P) \times \frac{1}{N} \sum_{i=1}^{N} f(x_i)
$$

Here’s how you can implement it in Python:

```python
import numpy as np
from scipy.spatial import ConvexHull

def monte_carlo_normalizing_constant(f, P, N=10000):
    """
    Computes the normalizing constant Z for a density f(x) supported on polytope P using Monte Carlo integration.
    
    Parameters:
        f : function
            The density function.
        P : ndarray
            An (M, d) array where each row is a vertex of the polytope.
        N : int
            Number of Monte Carlo samples.

    Returns:
        Z : float
            Approximate normalizing constant.
    """
    # Compute polytope volume using ConvexHull
    hull = ConvexHull(P)
    volume = hull.volume

    # Sample uniformly from the polytope using rejection sampling
    samples = np.random.uniform(P.min(axis=0), P.max(axis=0), size=(N, P.shape[1]))
    inside_polytope = np.all(np.dot(samples, hull.equations[:, :-1].T) + hull.equations[:, -1] <= 0, axis=1)

    valid_samples = samples[inside_polytope][:N]
    f_values = np.array([f(x) for x in valid_samples])
    
    return volume * np.mean(f_values)

# Example usage:
def density_function(x):
    return np.exp(-np.linalg.norm(x)**2)  # Example density function

# Define a polytope by its vertices (e.g., a square or simplex)
P = np.array([[0, 0], [1, 0], [1, 1], [0, 1]])  # A square polytope

Z_approx = monte_carlo_normalizing_constant(density_function, P, N=10000)
print("Estimated normalizing constant Z:", Z_approx)
```

#### **2. Exact Integration (Deterministic Approach)**
If \( f(x) \) is simple and \( P \) has a well-structured shape (e.g., a simplex, cube), exact integration using symbolic or numerical integration is possible. You can use `scipy.integrate.nquad`:

```python
from scipy.integrate import nquad

def exact_normalizing_constant(f, bounds):
    return nquad(f, bounds)[0]

# Example for a 2D polytope (rectangle with bounds)

Z_exact = exact_normalizing_constant(density_function, bounds)

```

This method is feasible only if integration bounds can be easily parameterized.

#### **3. Triangulation-Based Numerical Integration**
If the polytope is a convex hull in 2D or 3D, you can break it into simplices and integrate over them. This method is more accurate than Monte Carlo but computationally expensive.

Would you like me to implement a triangulation-based method for a general polytope? 🚀

The volume of a convex hull in `scipy.spatial.ConvexHull` is computed using **the sum of the signed volumes of simplices** that form the convex hull. The approach is based on the **determinant formula for the volume of a simplex**.

### **How is ConvexHull Volume Computed?**
1. **Triangulation into Simplices**:  
   - The convex hull is represented as a set of **simplices** (i.e., triangles in 2D, tetrahedra in 3D, and higher-dimensional generalizations).
   - These simplices are determined by the `simplices` attribute of `ConvexHull`, which provides the indices of points forming each simplex.

2. **Signed Volume of a Simplex**:
   - The volume of a \( d \)-dimensional simplex formed by \( d+1 \) points \( v_0, v_1, ..., v_d \) is given by:

     $$
     V = \frac{1}{d!} \left| \det \begin{bmatrix} v_1 - v_0 & v_2 - v_0 & \dots & v_d - v_0 \end{bmatrix} \right|
     $$

   - This determinant computes the **signed** volume of the simplex.

3. **Summing Over Simplices**:
   - The total volume of the convex hull is the sum of the absolute volumes of all simplices forming the hull.

---

### **How `scipy.spatial.ConvexHull` Computes It?**
In `scipy.spatial.ConvexHull`, the `volume` attribute automatically computes this using the determinant method:

```python
import numpy as np
from scipy.spatial import ConvexHull

# Example points defining a convex polytope in 3D
points = np.random.rand(10, 3)  # 10 points in 3D space

# Compute the convex hull
hull = ConvexHull(points)

# The computed volume of the convex hull
print("Volume of the ConvexHull:", hull.volume)
```

- `hull.volume` sums the volumes of all the simplices forming the convex hull.

---

### **Manual Computation of Volume**
If you want to manually compute the volume from the simplices, you can do:

```python
def compute_convex_hull_volume(points):
    hull = ConvexHull(points)
    total_volume = 0

    for simplex in hull.simplices:
        vertices = points[simplex]
        base_vertex = vertices[0]
        matrix = vertices[1:] - base_vertex  # Compute edge vectors
        simplex_volume = np.abs(np.linalg.det(matrix)) / np.math.factorial(len(vertices) - 1)
        total_volume += simplex_volume

    return total_volume

# Compute manually
manual_volume = compute_convex_hull_volume(points)
print("Manually computed volume:", manual_volume)
print("Scipy ConvexHull volume:", hull.volume)
```

---

### **Key Takeaways**
- **Uses triangulation**: Decomposes the convex hull into simplices.
- **Computes signed simplex volumes**: Uses determinant of the edge matrix.
- **Summation gives total volume**: The absolute values of simplex volumes are summed.

First, let us restate the forward transformation clearly.  We have the following definitions (in \(K\)-dimensional space):

1. A point in “ball coordinates” is
$$
\mathbf{x} \;=\; 
\begin{bmatrix}
r\\[6pt]
\mathbf{s}
\end{bmatrix}
\;\in\;\mathbb{B}^{K}(1),
$$
where 
- \(0 \,\le\, r \,\le\, 1\) is a scalar “radius” coordinate, 
- \(\mathbf{s}\in \mathbb{S}^{K-1}\) is a unit vector in \(\mathbb{R}^{K}\) (i.e., \(\|\mathbf{s}\|_2=1\)), and 
- \(\mathbb{B}^{K}(1)\) is the unit ball in \(\mathbb{R}^K\).

2. The corresponding “original” vector (in \(\mathbb{R}^K\)) is
$$
\mathbf{v}^\ddagger \;=\; d\,\mathbf{s}, 
\quad \text{where} \quad 
d \;=\; \|\mathbf{v}^\ddagger\|_2.
$$
However, from the context of your equations, we identify that
$$
d \;=\; r \,\alpha^{\max}(\mathbf{s}),
$$
where 
$$
\alpha^{\max}(\mathbf{s}) 
\;=\; \min\!\bigl(\mathbf{\alpha}\bigr)_{\ge0}, 
\quad
\text{with} 
\quad
\mathbf{\alpha}
\;=\;
\frac{\mathbf{b}^\ddagger}{\,\mathbf{A}^\ddagger\cdot \mathbf{s}\,}
\;=\;
\begin{bmatrix}
\frac{b_1^\ddagger}{(\mathbf{a}_1^\ddagger)^\mathsf{T}\mathbf{s}}\\[6pt]
\frac{b_2^\ddagger}{(\mathbf{a}_2^\ddagger)^\mathsf{T}\mathbf{s}}\\[4pt]
\vdots \\[4pt]
\frac{b_m^\ddagger}{(\mathbf{a}_m^\ddagger)^\mathsf{T}\mathbf{s}}
\end{bmatrix}
$$
and the notation \(\min(\mathbf{\alpha})_{\ge 0}\) indicates the smallest nonnegative component among these ratios (i.e., the limiting/active constraint).  Locally (for a given direction $\mathbf{s}$) we assume there is a single “active” constraint 
$$
i^* \;=\; \arg\!\min_{\,i\,:\,(\mathbf{a}_i^\ddagger)^\mathsf{T}\mathbf{s}>0}
\frac{b_i^\ddagger}{(\mathbf{a}_i^\ddagger)^\mathsf{T}\mathbf{s}}\ .
$$
Hence, **locally** near \(\mathbf{s}\), we have
$$
\alpha^{\max}(\mathbf{s})
\;=\;
\frac{b_{\,i^*}^\ddagger}{(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}}
\quad
\bigl(\text{assuming }(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}>0\bigr).
$$

Putting it all together, the **forward mapping** from \(\mathbf{x}=(r,\mathbf{s})\) to \(\mathbf{v}^\ddagger\) is:

\[
\mathbf{v}^\ddagger 
\;=\; 
d\,\mathbf{s} 
\;=\; 
\bigl[r\,\alpha^{\max}(\mathbf{s})\bigr]\,\mathbf{s}.
\]

---

## The Jacobian

We want the total derivative (Jacobian) of \(\mathbf{v}^\ddagger\) with respect to \(\mathbf{x}=(r,\mathbf{s})\).  Observe that:

\[
\mathbf{v}^\ddagger(r,\mathbf{s})
\;=\;
r \,\alpha^{\max}(\mathbf{s})\,\mathbf{s}.
\]

It is helpful to set 
\[
\alpha(\mathbf{s}) \;:=\; \alpha^{\max}(\mathbf{s}) 
\;=\; \frac{b_{i^*}^\ddagger}{(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}}
\]
locally (assuming the constraint \(i^*\) is the active one).  Then

\[
\mathbf{v}^\ddagger(r,\mathbf{s})
\;=\; 
r\,\alpha(\mathbf{s})\,\mathbf{s}.
\]

### 1. Partial derivative w.r.t. \(r\)

Treat \(\mathbf{s}\) as constant and differentiate w.r.t.\ \(r\).  By inspection,

\[
\frac{\partial \mathbf{v}^\ddagger}{\partial r}
\;=\;
\alpha(\mathbf{s})\,\mathbf{s}.
\]

This is a \(K\times 1\) column vector (since \(\mathbf{v}^\ddagger\in\mathbb{R}^K\) and \(r\) is a single scalar).

### 2. Partial derivative w.r.t. \(\mathbf{s}\)

Now treat \(r\) as constant and differentiate w.r.t.\ the vector \(\mathbf{s}\).  We use the product rule:

\[
\mathbf{v}^\ddagger(r,\mathbf{s})
\;=\;
r\,\bigl[\alpha(\mathbf{s})\,\mathbf{s}\bigr].
\]

Hence, for a small variation \(\delta \mathbf{s}\in\mathbb{R}^K\),

\[
\delta\mathbf{v}^\ddagger
\;=\;
r
\Bigl[
\delta \bigl(\alpha(\mathbf{s})\,\mathbf{s}\bigr)
\Bigr]
\;=\;
r
\Bigl[
\alpha(\mathbf{s})\,\delta\mathbf{s}
\;+\;
\mathbf{s}\,\delta\alpha(\mathbf{s})
\Bigr].
\]
Since \(\delta \alpha(\mathbf{s}) \;=\; \nabla_{\mathbf{s}} \alpha(\mathbf{s}) \;\cdot\; \delta\mathbf{s},\) we can write

\[
\delta\mathbf{v}^\ddagger
\;=\;
r
\Bigl[
\alpha(\mathbf{s})\,I 
\;+\; 
\mathbf{s}\,\bigl(\nabla_{\mathbf{s}}\alpha(\mathbf{s})\bigr)^\mathsf{T}
\Bigr]
\,\delta\mathbf{s}.
\]

Therefore, the \(K\times K\) matrix representing 
\(\frac{\partial \mathbf{v}^\ddagger}{\partial \mathbf{s}}\)
is

\[
\frac{\partial \mathbf{v}^\ddagger}{\partial \mathbf{s}}
\;=\;
r
\Bigl[
\alpha(\mathbf{s})\,I 
\;+\; 
\mathbf{s}\,\bigl(\nabla_{\mathbf{s}}\alpha(\mathbf{s})\bigr)^\mathsf{T}
\Bigr].
\]

#### Computing \(\nabla_{\mathbf{s}}\alpha(\mathbf{s})\)

Recall that, locally,
\[
\alpha(\mathbf{s}) 
\;=\;
\frac{b_{\,i^*}^\ddagger}{(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}}.
\]
Its gradient w.r.t.\ \(\mathbf{s}\) is

\[
\nabla_{\mathbf{s}} \alpha(\mathbf{s})
\;=\;
b_{\,i^*}^\ddagger \,\nabla_{\mathbf{s}}
\Bigl[\tfrac{1}{(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}}\Bigr]
\;=\;
b_{\,i^*}^\ddagger
\Bigl[
-\,\frac{\mathbf{a}_{\,i^*}^\ddagger}{\bigl((\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}\bigr)^2}
\Bigr]
\;=\;
-\,\frac{b_{\,i^*}^\ddagger}{\bigl((\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}\bigr)^2}
\;\mathbf{a}_{\,i^*}^\ddagger.
\]
But also note
\(\alpha(\mathbf{s})\,\bigl((\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}\bigr) = b_{\,i^*}^\ddagger.\)
Hence
\[
\frac{b_{\,i^*}^\ddagger}{\bigl((\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}\bigr)^2}
\;=\;
\frac{\alpha(\mathbf{s})}{(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}}
\quad\Longrightarrow\quad
\nabla_{\mathbf{s}}\alpha(\mathbf{s})
\;=\;
-\,\alpha(\mathbf{s})
\,\frac{\mathbf{a}_{\,i^*}^\ddagger}{(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}}.
\]
So
\[
\mathbf{s}\,\bigl(\nabla_{\mathbf{s}}\alpha(\mathbf{s})\bigr)^\mathsf{T}
\;=\;
-\,\alpha(\mathbf{s})\,\mathbf{s}\,
\frac{\bigl(\mathbf{a}_{\,i^*}^\ddagger\bigr)^\mathsf{T}}
{\bigl((\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}\bigr)}.
\]
Therefore,

\[
\frac{\partial \mathbf{v}^\ddagger}{\partial \mathbf{s}}
\;=\;
r\,\alpha(\mathbf{s})
\Bigl[
I 
\;-\;
\frac{\mathbf{s}\,\bigl(\mathbf{a}_{\,i^*}^\ddagger\bigr)^\mathsf{T}}%
     {\bigl((\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}\bigr)}
\Bigr].
\]

Putting it all together:

- **Partial w.r.t.** \(r\):
  \[
  \frac{\partial \mathbf{v}^\ddagger}{\partial r}
  \;=\;
  \alpha(\mathbf{s})\,\mathbf{s}.
  \]

- **Partial w.r.t.** \(\mathbf{s}\):
  \[
  \frac{\partial \mathbf{v}^\ddagger}{\partial \mathbf{s}}
  \;=\;
  r\,\alpha(\mathbf{s})
  \Bigl[
  I 
  \;-\;
  \frac{\mathbf{s}\,\bigl(\mathbf{a}_{\,i^*}^\ddagger\bigr)^\mathsf{T}}%
       {\bigl((\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}\bigr)}
  \Bigr].
  \]

---

## Block-Matrix Form of the Jacobian

Since
\(\mathbf{x}=(r,\mathbf{s})\in \mathbb{R}\times \mathbb{R}^K\)
(\(\mathbf{s}\) is a \(K\)-vector on the unit sphere, albeit typically you only have \(K-1\) degrees of freedom there, but formally you can write the differential in \(\mathbb{R}^K\) if you wish), the **Jacobian** 
\(\tfrac{\partial \mathbf{v}^\ddagger}{\partial \mathbf{x}}\)
is a \(K\times (1+K)\) matrix.  It can be written in “block” form as

\[
\frac{\partial \mathbf{v}^\ddagger}{\partial \mathbf{x}}
\;=\;
\Bigl[
\underbrace{\frac{\partial \mathbf{v}^\ddagger}{\partial r}}_{\displaystyle K\times 1}
\;\Big\vert\;
\underbrace{\frac{\partial \mathbf{v}^\ddagger}{\partial \mathbf{s}}}_{\displaystyle K\times K}
\Bigr].
\]

Substituting the expressions above,

1. **First block column** (w.r.t.\ \(r\)):

   \[
   \frac{\partial \mathbf{v}^\ddagger}{\partial r}
   \;=\;
   \alpha(\mathbf{s})\,\mathbf{s}.
   \]

2. **Second block** (w.r.t.\ \(\mathbf{s}\)):

   \[
   \frac{\partial \mathbf{v}^\ddagger}{\partial \mathbf{s}}
   \;=\;
   r\,\alpha(\mathbf{s})
   \Bigl[
   I 
   \;-\;
   \frac{\mathbf{s}\,\bigl(\mathbf{a}_{\,i^*}^\ddagger\bigr)^\mathsf{T}}%
        {\bigl((\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}\bigr)}
   \Bigr].
   \]

Hence, as a matrix:

\[
\boxed{
\frac{\partial \mathbf{v}^\ddagger}{\partial \mathbf{x}}
\;=\;
\begin{bmatrix}
\alpha(\mathbf{s})\,\mathbf{s} 
& 
r\,\alpha(\mathbf{s})
\Bigl[
I 
- 
\dfrac{\mathbf{s}\,\bigl(\mathbf{a}_{\,i^*}^\ddagger\bigr)^\mathsf{T}}
      {(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}}
\Bigr]
\end{bmatrix}
}
\]
where
\(\alpha(\mathbf{s}) \;=\; \dfrac{b_{\,i^*}^\ddagger}{(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}}\).

---

## Elementwise Form

If you prefer each element explicitly, label the components of 
\(\mathbf{v}^\ddagger = \bigl(v_1^\ddagger,\dots,v_K^\ddagger\bigr)^\mathsf{T}\)
and the coordinates 
\(\mathbf{x} = (r,s_1,\dots,s_K)\) (treating \(\mathbf{s}\) just as a \(K\)-vector). Then:

1. **Partial w.r.t.** \(r\) (this is the first column):
   \[
   \frac{\partial v_i^\ddagger}{\partial r}
   \;=\;
   \alpha(\mathbf{s})\,s_i,
   \quad
   \text{for }i=1,\dots,K.
   \]

2. **Partial w.r.t.** \(s_j\) (the \((j+1)\)-th column):
   \[
   \frac{\partial v_i^\ddagger}{\partial s_j}
   \;=\;
   r\,\alpha(\mathbf{s})
   \biggl[
   \delta_{ij}
   \;-\;
   \frac{s_i\,\bigl(a_{\,i^*}^\ddagger\bigr)_j}%
        {(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}}
   \biggr],
   \]
   where \(\delta_{ij}\) is the Kronecker delta (1 if \(i=j\), 0 otherwise), and 
   \(\bigl(a_{\,i^*}^\ddagger\bigr)_j\) is the \(j\)-th component of the row-vector 
   \(\mathbf{a}_{\,i^*}^\ddagger\).

That fully specifies each entry of the Jacobian in a standard coordinate form.

# Final Summary

\[
\mathbf{v}^\ddagger 
\;=\; 
r\,\alpha^{\max}(\mathbf{s})\,\mathbf{s},
\quad
\text{with}
\quad
\alpha^{\max}(\mathbf{s})
\;=\;
\min_{\,i\,:\,(\mathbf{a}_i^\ddagger)^\mathsf{T}\mathbf{s}>0}
\frac{b_i^\ddagger}{(\mathbf{a}_i^\ddagger)^\mathsf{T}\mathbf{s}}\ .
\]
Locally (with \(i^*\) the “active” constraint),

\[
\alpha(\mathbf{s})
\;=\;
\frac{b_{\,i^*}^\ddagger}{(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}}.
\]

Hence the Jacobian \(\displaystyle \frac{\partial \mathbf{v}^\ddagger}{\partial(r,\mathbf{s})}\) in block form is

\[
\boxed{
\frac{\partial \mathbf{v}^\ddagger}{\partial \mathbf{x}}
\;=\;
\begin{bmatrix}
\alpha(\mathbf{s})\,\mathbf{s} 
& 
r\,\alpha(\mathbf{s})
\Bigl[
I 
- 
\dfrac{\mathbf{s}\,\bigl(\mathbf{a}_{\,i^*}^\ddagger\bigr)^\mathsf{T}}
      {(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}}
\Bigr]
\end{bmatrix}
\quad
\text{(\(K\times (1+K)\) matrix).}
}
\]

And in elementwise form:

- For \(i=1,\dots,K\):
  \[
  \frac{\partial v_i^\ddagger}{\partial r}
  \;=\;
  \alpha(\mathbf{s})\,s_i.
  \]
- For \(i=1,\dots,K\) and \(j=1,\dots,K\):
  \[
  \frac{\partial v_i^\ddagger}{\partial s_j}
  \;=\;
  r\,\alpha(\mathbf{s})
  \Bigl[
    \delta_{ij}
    \;-\;
    \frac{s_i\,\bigl(\mathbf{a}_{\,i^*}^\ddagger\bigr)_j}%
         {(\mathbf{a}_{\,i^*}^\ddagger)^\mathsf{T}\mathbf{s}}
  \Bigr].
  \]

This completes the specification of all entries of the Jacobian for arbitrary dimension \(K\).

````python
import numpy as np

def compute_jacobian(A, b, s, r, active_constraint_index):
    """
    Compute the Jacobian matrix J = dv/dx for the transformation from the unit ball to the polytope.

    Parameters:
    - A: numpy.ndarray of shape (M, K)
        Constraint matrix defining the polytope (A @ v <= b).
    - b: numpy.ndarray of shape (M,)
        Constraint bounds.
    - s: numpy.ndarray of shape (K,)
        Direction vector on the unit sphere (||s||_2 = 1).
    - r: float
        Radial component within the unit ball (0 <= r <= 1).
    - active_constraint_index: int
        Index of the active constraint (0-based indexing).

    Returns:
    - J: numpy.ndarray of shape (K, 1 + K)
        Jacobian matrix of the transformation.
    """
    # Validate inputs
    M, K = A.shape
    assert b.shape == (M,), "Vector b must have shape (M,)"
    assert s.shape == (K,), "Vector s must have shape (K,)"
    assert np.isclose(np.linalg.norm(s), 1), "Vector s must be a unit vector."
    assert 0 <= active_constraint_index < M, "Invalid active_constraint_index."

    # Extract the active constraint
    A_i = A[active_constraint_index, :]  # Shape: (K,)

    # Compute dot product A_i . s
    A_i_dot_s = np.dot(A_i, s)
    if A_i_dot_s == 0:
        raise ValueError("Dot product of active constraint and s is zero, division by zero encountered.")

    # Compute alpha_max
    alpha_max = b[active_constraint_index] / A_i_dot_s

    # Compute the first column of J: alpha_max * s
    first_col = alpha_max * s  # Shape: (K,)

    # Compute the term (s outer A_i) / (A_i . s)
    print(s)
    print(A_i)
    outer_term = np.outer(s, A_i) / A_i_dot_s  # Shape: (K, K)

    # Compute (I - outer_term)
    identity = np.eye(K)
    I_minus_outer = identity - outer_term  # Shape: (K, K)

    # Compute the remaining columns: r * alpha_max * (I - outer_term)
    remaining_cols = r * alpha_max * I_minus_outer  # Shape: (K, K)

    # Assemble the Jacobian matrix J by concatenating first_col and remaining_cols
    # Reshape first_col to (K, 1) for horizontal stacking
    J = np.hstack((first_col.reshape(K, 1), remaining_cols))  # Shape: (K, 1 + K)

    return J

# Example Usage
if __name__ == "__main__":
    # Define polytope constraints (Example: 2D polytope)
    A = np.array([
        [1, 0],
        [0, 1],
        [-1, 0],
        [0, -1]
    ])  # Shape: (4, 2)

    b = np.array([2, 2, 1, 1])  # Shape: (4,)

    # Define unit ball coordinates
    r = 0.5
    s = np.array([np.sqrt(0.5), np.sqrt(0.5)])  # Unit vector in 2D

    # Active constraint index (e.g., first constraint is active)
    active_constraint_index = 0  # 0-based indexing

    # Compute Jacobian
    J = compute_jacobian(A, b, s, r, active_constraint_index)

    # Display the Jacobian
    np.set_printoptions(precision=4, suppress=True)
    print("Jacobian Matrix J:")
    print(J)
````

````python
Certainly! Let's compute the partial derivative \(\frac{\partial v_1}{\partial s_1}\) within the context of the transformation you've provided. We'll approach this step-by-step to ensure clarity and thoroughness.

### **Recap of the Transformation**

From your initial transformation, the mapping from the unit ball coordinates \(\vec{x} = \begin{bmatrix} r \\ \vec{s} \end{bmatrix}\) to the polytope coordinates \(\vec{v}^\ddagger\) is given by:

\[
\vec{v}^\ddagger = \mathscr{b}^{-1}(\vec{x}) = r \cdot \alpha^{\max} \cdot \vec{s}
\]

Where:
- \( r \in [0,1] \) is the radial component.
- \( \vec{s} \in \mathbb{S}^{K-1}(1) \) is the normalized direction vector (i.e., \(\|\vec{s}\|_2 = 1\)).
- \( \alpha^{\max} \) is determined by the active constraint, specifically:
  
  \[
  \alpha^{\max} = \frac{b_i}{\pmb{A}_i \cdot \vec{s}}
  \]
  
  Here, \( \pmb{A}_i \) is the \(i\)-th row of the constraint matrix \( \pmb{A} \), and \( b_i \) is the corresponding bound for the active constraint.

### **Objective**

Compute the partial derivative of the first component of \(\vec{v}^\ddagger\) with respect to \(s_1\):

\[
\frac{\partial v_1}{\partial s_1}
\]

### **Step-by-Step Derivation**

#### **1. Express \(v_1\) Explicitly**

Given the inverse mapping:

\[
v_1 = r \cdot \alpha^{\max} \cdot s_1
\]

Substituting the expression for \(\alpha^{\max}\):

\[
v_1 = r \cdot \left( \frac{b_i}{\pmb{A}_i \cdot \vec{s}} \right ) \cdot s_1 = \frac{r b_i s_1}{\pmb{A}_i \cdot \vec{s}}
\]

#### **2. Apply the Chain Rule**

To find \(\frac{\partial v_1}{\partial s_1}\), we'll differentiate \(v_1\) with respect to \(s_1\). Notice that \(v_1\) depends on \(s_1\) both explicitly and implicitly through \(\alpha^{\max}\).

\[
v_1 = r \cdot \alpha^{max} \cdot s_1
\]

Thus, using the product rule:

\[
\frac{\partial v_1}{\partial s_1} = r \cdot \frac{\partial \alpha^{\max}}{\partial s_1} \cdot s_1 + r \cdot \alpha^{\max} \cdot \frac{\partial s_1}{\partial s_1}
\]

Simplifying:

\[
\frac{\partial v_1}{\partial s_1} = r \cdot s_1 \cdot \frac{\partial \alpha^{\max}}{\partial s_1} + r \cdot \alpha^{\max} \cdot 1
\]

\[
\frac{\partial v_1}{\partial s_1} = r \cdot s_1 \cdot \frac{\partial \alpha^{\max}}{\partial s_1} + r \cdot \alpha^{\max}
\]

#### **3. Compute \(\frac{\partial \alpha^{\max}}{\partial s_1}\)**

From your previous transformation steps, we have:

\[
\alpha^{\max} = \frac{b_i}{\pmb{A}_i \cdot \vec{s}} = \frac{b_i}{A_{i1} s_1 + A_{i2} s_2 + \dots + A_{iK} s_K}
\]

Let \( D = \pmb{A}_i \cdot \vec{s} = \sum_{j=1}^K A_{ij} s_j \). Then:

\[
\alpha^{\max} = \frac{b_i}{D}
\]

Differentiating with respect to \(s_1\):

\[
\frac{\partial \alpha^{\max}}{\partial s_1} = \frac{\partial}{\partial s_1} \left( \frac{b_i}{D} \right ) = -\frac{b_i}{D^2} \cdot \frac{\partial D}{\partial s_1}
\]

\[
\frac{\partial D}{\partial s_1} = A_{i1}
\]

Thus:

\[
\frac{\partial \alpha^{\max}}{\partial s_1} = -\frac{b_i \cdot A_{i1}}{(\pmb{A}_i \cdot \vec{s})^2}
\]

#### **4. Substitute \(\frac{\partial \alpha^{\max}}{\partial s_1}\) Back into \(\frac{\partial v_1}{\partial s_1}\)**

Plugging the expression for \(\frac{\partial \alpha^{\max}}{\partial s_1}\) back into the derivative of \(v_1\):

\[
\frac{\partial v_1}{\partial s_1} = r \cdot s_1 \cdot \left( -\frac{b_i \cdot A_{i1}}{(\pmb{A}_i \cdot \vec{s})^2} \right ) + r \cdot \frac{b_i}{\pmb{A}_i \cdot \vec{s}}
\]

Simplify the expression:

\[
\frac{\partial v_1}{\partial s_1} = -\frac{r b_i A_{i1} s_1}{(\pmb{A}_i \cdot \vec{s})^2} + \frac{r b_i}{\pmb{A}_i \cdot \vec{s}}
\]

Factor out common terms:

\[
\frac{\partial v_1}{\partial s_1} = \frac{r b_i}{(\pmb{A}_i \cdot \vec{s})^2} \left( -A_{i1} s_1 + \pmb{A}_i \cdot \vec{s} \right )
\]

Recall that \( \pmb{A}_i \cdot \vec{s} = A_{i1} s_1 + A_{i2} s_2 + \dots + A_{iK} s_K \), so:

\[
-A_{i1} s_1 + \pmb{A}_i \cdot \vec{s} = A_{i2} s_2 + \dots + A_{iK} s_K = \sum_{j=2}^K A_{ij} s_j
\]

However, to maintain a general form, we can leave it as is:

\[
\frac{\partial v_1}{\partial s_1} = \frac{r b_i}{(\pmb{A}_i \cdot \vec{s})^2} \left( \pmb{A}_i \cdot \vec{s} - A_{i1} s_1 \right )
\]

But simplifying further, we recognize that:

\[
\pmb{A}_i \cdot \vec{s} - A_{i1} s_1 = \sum_{j=2}^K A_{ij} s_j
\]

However, for the purpose of expressing \(\frac{\partial v_1}{\partial s_1}\) in a compact and meaningful way, we can factor back to the earlier step:

\[
\frac{\partial v_1}{\partial s_1} = \frac{r b_i}{\pmb{A}_i \cdot \vec{s}} \left( 1 - \frac{A_{i1} s_1}{\pmb{A}_i \cdot \vec{s}} \right )
\]

### **Final Expression**

Thus, the partial derivative of \(v_1\) with respect to \(s_1\) is:

\[
\boxed{ \frac{\partial v_1}{\partial s_1} = \frac{r b_i}{\pmb{A}_i \cdot \vec{s}} \left( 1 - \frac{A_{i1} s_1}{\pmb{A}_i \cdot \vec{s}} \right ) }
\]

### **Interpretation**

- **Scaling Factor \(\frac{r b_i}{\pmb{A}_i \cdot \vec{s}}\)**: This term scales the derivative based on the radial component \(r\), the bound \(b_i\), and the dot product \(\pmb{A}_i \cdot \vec{s}\).

- **Adjustment Term \(\left( 1 - \frac{A_{i1} s_1}{\pmb{A}_i \cdot \vec{s}} \right )\)**: This term accounts for the influence of the active constraint's coefficient \(A_{i1}\) and the direction component \(s_1\) relative to the overall scaling determined by \(\pmb{A}_i \cdot \vec{s}\).

### **Verification via Jacobian Matrix**

Recall from the **Fully Expanded Jacobian Matrix** in our previous discussion, the general expression for the Jacobian \(\mathbf{J} = \frac{\partial \vec{v}}{\partial \vec{x}}\) is:

\[
\mathbf{J} = \begin{bmatrix} 
\alpha^{\max} s_1 & r \cdot \frac{b_i}{\pmb{A}_i \cdot \vec{s}} \left( 1 - \frac{s_1 A_{i1}}{\pmb{A}_i \cdot \vec{s}} \right ) & \dots \\
\alpha^{\max} s_2 & \dots & \dots \\
\vdots & \vdots & \ddots \\
\alpha^{\max} s_K & \dots & \dots 
\end{bmatrix}
\]

Specifically, for \(\frac{\partial v_1}{\partial s_1}\), the entry is:

\[
\frac{\partial v_1}{\partial s_1} = r \cdot \frac{b_i}{\pmb{A}_i \cdot \vec{s}} \left( 1 - \frac{s_1 A_{i1}}{\pmb{A}_i \cdot \vec{s}} \right )
\]

Which matches our derived expression.

### **Conclusion**

The partial derivative \(\frac{\partial v_1}{\partial s_1}\) quantifies how the first component of the polytope point \(\vec{v}^\ddagger\) changes in response to infinitesimal changes in the first component \(s_1\) of the unit ball coordinate \(\vec{s}\). This derivative is essential for understanding the sensitivity and behavior of the transformation, especially in optimization and geometric analysis contexts.
````

To compute the Jacobian matrix \(\frac{\partial \vec{v}^\ddagger}{\partial \vec{p}}\) of the transformation \(\mathscr{b}^{-1}\), let's break down the relationships and derive the partial derivatives step by step.

### Given Transformations

1. **Inverse Mapping:**
   \[
   \vec{v}^\ddagger = \mathscr{b}^{-1}(\vec{v}^{\, \mathcal{S}}) = r \cdot \alpha^{\max} \cdot \vec{s}
   \]
   
2. **Polar Coordinates:**
   \[
   \vec{p} = \polar(r, \vec{s}) = \left[ r, s_1, s_2, \dots, s_{K-1}, \vartheta \right]^T
   \]
   where \(\vartheta = \atan2(s_K, s_{K+1})\).

### Jacobian Matrix Structure

The Jacobian matrix \(\frac{\partial \vec{v}^\ddagger}{\partial \vec{p}}\) captures how each component of \(\vec{v}^\ddagger\) changes with respect to each component of \(\vec{p}\). Assuming \(\vec{p}\) consists of \(r\), \(s_1, s_2, \dots, s_{K-1}\), and \(\vartheta\), the Jacobian will have the following structure:

\[
\frac{\partial \vec{v}^\ddagger}{\partial \vec{p}} = 
\begin{bmatrix}
\frac{\partial v_1}{\partial r} & \frac{\partial v_1}{\partial s_1} & \cdots & \frac{\partial v_1}{\partial s_{K-1}}} & \frac{\partial v_1}{\partial \vartheta} \\
\frac{\partial v_2}{\partial r} & \frac{\partial v_2}{\partial s_1} & \cdots & \frac{\partial v_2}{\partial s_{K-1}}} & \frac{\partial v_2}{\partial \vartheta} \\
\vdots & \vdots & \ddots & \vdots & \vdots \\
\frac{\partial v_K}{\partial r} & \frac{\partial v_K}{\partial s_1} & \cdots & \frac{\partial v_K}{\partial s_{K-1}}} & \frac{\partial v_K}{\partial \vartheta} \\
\end{bmatrix}
\]

### Deriving Partial Derivatives

1. **Partial Derivative with Respect to \(r\):**
   \[
   \frac{\partial \vec{v}^\ddagger}{\partial r} = \alpha^{\max} \cdot \vec{s}
   \]
   For each component \(v_i\):
   \[
   \frac{\partial v_i}{\partial r} = \alpha^{\max} \cdot s_i
   \]

2. **Partial Derivative with Respect to \(s_j\) (for \(j = 1, 2, \dots, K-1\)):**
   \[
   \frac{\partial \vec{v}^\ddagger}{\partial s_j} = r \cdot \left( \frac{\partial \alpha^{\max}}{\partial s_j} \cdot \vec{s} + \alpha^{\max} \cdot \vec{e}_j \right)
   \]
   where \(\vec{e}_j\) is the unit vector in the \(j\)-th direction.

   For each component \(v_i\):
   \[
   \frac{\partial v_i}{\partial s_j} = r \cdot \left( \frac{\partial \alpha^{\max}}{\partial s_j} \cdot s_i + \alpha^{\max} \cdot \delta_{ij} \right)
   \]
   Here, \(\delta_{ij}\) is the Kronecker delta, which is 1 if \(i = j\) and 0 otherwise.

3. **Partial Derivative with Respect to \(\vartheta\):**
   \[
   \frac{\partial \vec{v}^\ddagger}{\partial \vartheta} = r \cdot \left( \frac{\partial \alpha^{\max}}{\partial \vartheta} \cdot \vec{s} + \alpha^{\max} \cdot \frac{\partial \vec{s}}{\partial \vartheta} \right)
   \]
   
   For each component \(v_i\):
   \[
   \frac{\partial v_i}{\partial \vartheta} = r \cdot \left( \frac{\partial \alpha^{\max}}{\partial \vartheta} \cdot s_i + \alpha^{\max} \cdot \frac{\partial s_i}{\partial \vartheta} \right)
   \]

### Summary of Jacobian Elements

Putting it all together, the Jacobian matrix elements are:

\[
\frac{\partial v_i}{\partial p_j} =
\begin{cases}
\alpha^{\max} \cdot s_i & \text{if } p_j = r \\
r \cdot \left( \frac{\partial \alpha^{\max}}{\partial s_j} \cdot s_i + \alpha^{\max} \cdot \delta_{ij} \right) & \text{if } p_j = s_j \\
r \cdot \left( \frac{\partial \alpha^{\max}}{\partial \vartheta} \cdot s_i + \alpha^{\max} \cdot \frac{\partial s_i}{\partial \vartheta} \right) & \text{if } p_j = \vartheta \\
\end{cases}
\]

### Notes on \(\frac{\partial \alpha^{\max}}{\partial s_j}\) and \(\frac{\partial \alpha^{\max}}{\partial \vartheta}\)

The terms \(\frac{\partial \alpha^{\max}}{\partial s_j}\) and \(\frac{\partial \alpha^{\max}}{\partial \vartheta}\) depend on how \(\alpha^{\max}\) is defined in terms of \(\vec{s}\) and \(\vartheta\). Specifically, \(\alpha^{\max}\) is given by:

\[
\alpha^{\max} = \min \left( \frac{b_i}{\pmb{A}_i \cdot \vec{s}} \right) \quad \text{for all } i \text{ such that } \frac{b_i}{\pmb{A}_i \cdot \vec{s}} \geq 0
\]

Assuming \(\alpha^{\max}\) is determined by a particular constraint \(i^*\) that achieves the minimum, the derivatives can be further specified as:

\[
\frac{\partial \alpha^{\max}}{\partial s_j} = -\frac{b_{i^*} \cdot A_{i^*j}}{(\pmb{A}_{i^*} \cdot \vec{s})^2}
\]
\[
\frac{\partial \alpha^{\max}}{\partial \vartheta} = -\frac{b_{i^*} \cdot \sum_{j=1}^K A_{i^*j} \cdot \frac{\partial s_j}{\partial \vartheta}}{(\pmb{A}_{i^*} \cdot \vec{s})^2}
\]

### Final Jacobian Expression

Combining all the components, the Jacobian matrix \(\frac{\partial \vec{v}^\ddagger}{\partial \vec{p}}\) is:

\[
\frac{\partial \vec{v}^\ddagger}{\partial \vec{p}} =
\begin{bmatrix}
\alpha^{\max} s_1 & r \left( \frac{\partial \alpha^{\max}}{\partial s_1} s_1 + \alpha^{\max} \right) & \cdots & r \left( \frac{\partial \alpha^{\max}}{\partial \vartheta} s_1 + \alpha^{\max} \frac{\partial s_1}{\partial \vartheta} \right) \\
\alpha^{\max} s_2 & r \left( \frac{\partial \alpha^{\max}}{\partial s_1} s_2 \right) & \cdots & r \left( \frac{\partial \alpha^{\max}}{\partial \vartheta} s_2 + \alpha^{\max} \frac{\partial s_2}{\partial \vartheta} \right) \\
\vdots & \vdots & \ddots & \vdots \\
\alpha^{\max} s_K & r \left( \frac{\partial \alpha^{\max}}{\partial s_{K-1}} s_K \right) & \cdots & r \left( \frac{\partial \alpha^{\max}}{\partial \vartheta} s_K + \alpha^{\max} \frac{\partial s_K}{\partial \vartheta} \right) \\
\end{bmatrix}
\]

### Conclusion

The Jacobian \(\frac{\partial \vec{v}^\ddagger}{\partial \vec{p}}\) encapsulates how each component of \(\vec{v}^\ddagger\) changes with respect to the polar coordinates \(r\), \(\vec{s}\), and \(\vartheta\). The exact expressions depend on the specific form of \(\alpha^{\max}\) and how it interacts with \(\vec{s}\) and \(\vartheta\).

**Final Expressions:**

\[

\]

These expressions provide the necessary elements to construct the Jacobian matrix for the given transformation.
